```python
import matplotlib.pyplot as plt
import torch
from src import *
```

```python
STEP_SIZE = 0.01
FIG_SIZE = (5,5)
CMAP = 'inferno'
ALPHA = 1
SHADING = 'nearest'
MARKER_T = '+'
MARKER_C = 'r'
```

```python
a = torch.arange(-5,5+STEP_SIZE,STEP_SIZE, dtype=t.float32)
b = torch.arange(-5,5+STEP_SIZE,STEP_SIZE, dtype=t.float32)
grid = torch.stack(torch.meshgrid(a, b), -1)
flat_grid = torch.reshape(grid, (-1,2))
```

```python
f01 = create_f01(2, seed=42)
results = torch.reshape(f01(flat_grid), grid.shape[:-1])

plt.figure(figsize=FIG_SIZE)
plt.pcolormesh(a, b, results, cmap=CMAP, alpha=ALPHA, shading=SHADING)
plt.scatter(*f01.x_opt.tolist()[::-1], marker=MARKER_T, c=MARKER_C)
plt.show()
```
