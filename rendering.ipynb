```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import torch
from src import *
```

```python
STEP_SIZE = 0.01
FIG_SIZE = (5,5)
CMAP = 'inferno'
ALPHA = 1
SHADING = 'nearest'
MARKER_T = '+'
MARKER_C = 'r'
```

```python
a = torch.arange(-5,5+STEP_SIZE,STEP_SIZE, dtype=t.float32)
b = torch.arange(-5,5+STEP_SIZE,STEP_SIZE, dtype=t.float32)
grid = torch.stack(torch.meshgrid(a, b), -1)
flat_grid = torch.reshape(grid, (-1,2))
```

```python
create_fn = [
	create_f01,
	create_f02,
	create_f03,
	create_f04,
	create_f05,
	create_f06,
	create_f07,
	create_f08,
	create_f09,
	create_f10,
	create_f11,
	create_f12,
	create_f13,
	create_f14,
]

name_fn = [
	'f01',
	'f02',
	'f03',
	'f04',
	'f05',
	'f06',
	'f07',
	'f08',
	'f09',
	'f10',
	'f11',
	'f12',
	'f13',
	'f14',
]
```

```python
for cr, name in zip(create_fn, name_fn):
	fn = cr(2, seed=42)
	results = torch.reshape(fn(flat_grid), grid.shape[:-1])
	results -= results.min() - 1

	plt.figure(figsize=FIG_SIZE)
	plt.pcolormesh(a, b, results, cmap=CMAP, alpha=ALPHA, shading=SHADING)
	plt.scatter(*fn.x_opt.tolist()[::-1], marker=MARKER_T, c=MARKER_C)
	plt.title(name)
	plt.show()

	plt.figure(figsize=FIG_SIZE)
	plt.pcolormesh(a, b, results, cmap=CMAP, alpha=ALPHA, shading=SHADING, norm=colors.LogNorm(results.min(), results.max()))
	plt.scatter(*fn.x_opt.tolist()[::-1], marker=MARKER_T, c=MARKER_C)
	plt.title("Log " + name)
	plt.show()
```
